# Breast Cancer Winsconsin

IN this notebook we will be using the Breast Cancer Wisconsin (Diagnostic) Database to create a classifier that can help diagnose patients. First, read through the description of the dataset (below).


### Setup

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt

%matplotlib inline


cancer = load_breast_cancer()

print(cancer.DESCR) # Print the data set description
```

### Examining the data

```python
cancer.keys() #Data keys
```

How many features does the breast cancer dataset have?

```python
len(cancer['feature_names']) 
```

Scikit-learn works with lists, numpy arrays, scipy-sparse matrices, and pandas DataFrames, so converting the dataset to a DataFrame is not necessary for training this model. Using a DataFrame does however help make many things easier such as munging data, so let's practice creating a classifier with a pandas DataFrame. 

```python
data = np.insert(cancer['data'], 30, values=cancer['target'], axis=1)
columns = np.insert(cancer['feature_names'], 30, 'target')
cancerdf = pd.DataFrame(data=data , index= pd.RangeIndex(start=0, stop=569, step=1), columns =columns)
    
cancerdf
```

What is the class distribution? (i.e. how many instances of `malignant` (encoded 0) and how many `benign` (encoded 1)?)

```python
target = cancerdf['target'].value_counts().sort_index()
target.index = cancer.target_names
target
```

### Create train-test split

```python
from sklearn.model_selection import train_test_split
X =cancerdf.iloc[:, 0:30]
y =cancerdf.iloc[:,30]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

### Create classifier object

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train, y_train)
```

### Train the classifier (fit the estimator) using the training data

Using our knn classifier to predict the class label using the mean value for each feature.

```python
means = cancerdf.mean()[:-1].values.reshape(1, -1) 
#gets the mean value for each feature and ignores the target column
#reshapes the data from 1 dimension to 2 (necessary for the precict method of KNeighborsClassifier).
knn.predict(means)
```

### Use the trained k-NN classifier model to classify new, previously unseen objects

```python
knn.predict(X_test)
```

### Estimate the accuracy of the classifier on future data, using the test data

```python
knn.score(X_test, y_test)
```

### Visualize the differet predicition scores between training and test sets, as well as malignant and benign cells.

```python
# Find the training and testing accuracies by target value (i.e. malignant, benign)
mal_train_X = X_train[y_train==0]
mal_train_y = y_train[y_train==0]
ben_train_X = X_train[y_train==1]
ben_train_y = y_train[y_train==1]

mal_test_X = X_test[y_test==0]
mal_test_y = y_test[y_test==0]
ben_test_X = X_test[y_test==1]
ben_test_y = y_test[y_test==1]

scores = [knn.score(mal_train_X, mal_train_y), knn.score(ben_train_X, ben_train_y), 
              knn.score(mal_test_X, mal_test_y), knn.score(ben_test_X, ben_test_y)]


plt.figure()

# Plot the scores as a bar chart
bars = plt.bar(np.arange(4), scores, color=['#4c72b0','#4c72b0','#55a868','#55a868'])

# directly label the score onto the bars
for bar in bars:
    height = bar.get_height()
    plt.gca().text(bar.get_x() + bar.get_width()/2, height*.90, '{0:.{1}f}'.format(height, 2), ha='center', color='w', fontsize=11)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

# remove the frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.xticks([0,1,2,3], ['Malignant\nTraining', 'Benign\nTraining', 'Malignant\nTest', 'Benign\nTest'], alpha=0.8);
plt.title('Training and Test Accuracies for Malignant and Benign Cells', alpha=0.8)
```
